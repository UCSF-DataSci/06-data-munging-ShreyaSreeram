```python
# Importing necessary libraries
import pandas as pd

data = pd.read_csv('messy_population_data.csv')

# Display basic information about the dataset
print("Basic Information:")
print(data.info())

# Descriptive Stats:
print("\nDescriptive Statistics:")
print(data.describe())

# Missing values in each columns:
print("\nMissing Values:")
missing_values = data.isnull().sum()
print(missing_values)

# Duplicate rows: 
print("\nDuplicate Rows:")
duplicate_rows = data.duplicated().sum()
print(f'Total duplicate rows: {duplicate_rows}')

# Display the number of unique values in each column
print("\nUnique Values per Column:")
unique_values = data.nunique()
print(unique_values)

print("\nValue Counts for 'income_groups':")
print(data['income_groups'].value_counts())

print("\nValue Counts for 'gender':")
print(data['gender'].value_counts())

print("\nYear Value Counts:")
print(data['year'].value_counts().sort_index())

print("\nPopulation Value Counts:")
print(data['population'].value_counts().sort_index())
```
